```python
import os
import shutil

##
# 查看 git 情况
##


path = 'C:\\Users\\16172\\Documents\\notfound945.github.io'
os.chdir(path)
switchdir = os.popen('git status')

print(os.popen('dir').read())

print(switchdir.read())
```

```python
##
# 清除原有文件
##


dst_root = 'C:\\Users\\16172\\Documents\\notfound945.github.io'
dst_js = dst_root + '\\js'
dst_css = dst_root + '\\css'
dst_fonts = dst_root + '\\fonts'
dst_icons = dst_root + '\\icons'
dst_index = dst_root + '\\index.html'
# dst_favicon = root + '\\favicon.png'
# dst_readme = root + '\\READEME.md'
# dst_license = root + '\\LICENSE'
# dst_ipynb = root + '\\soup_parser.ipynb'

#os.remove(path)   #删除文件
# os.removedirs(path)   #删除空文件夹
# os.rmdir(path)    #删除空文件夹
# shutil.rmtree(path)    #递归删除文件夹，即：删除非空文件夹

try:
    shutil.rmtree(dst_js)
    shutil.rmtree(dst_css)
    shutil.rmtree(dst_fonts)
    shutil.rmtree(dst_icons)
    os.remove(dst_index)
except shutil.Error as e:
    for src, dst, msg in e.args[0]:
         # src is source name
         # dst is destination name
         # msg is error message from exception
         print(dst, src, msg)
```

```python
##
# 提交 commit 使其更改
##
# git add . ：他会监控工作区的状态树，使用它会把工作时的所有变化提交到暂存区
# 包括文件内容修改(modified)以及新文件(new)，但不包括被删除的文件
# git add -u ：他仅监控已经被add的文件（即tracked file）
# 他会将被修改的文件提交到暂存区。add -u 不会提交新文件（untracked file）
# （git add --update的缩写）
# git add -A ：是上面两个功能的合集（git add --all的缩写）
#


path = 'C:\\Users\\16172\\Documents\\notfound945.github.io'
os.chdir(path)
pull = os.popen('git pull')
print(pull.read())
status = os.popen('git status')
print(status.read())
gitadd = os.popen('git add -A')
print(gitadd.read())
commit = os.popen('git commit -m  "To trigger the change of git."')
print(commit.read())
```

```python
##
# 复制更新文件到 github.io
##

src_root = 'C:\\Users\\16172\\Documents\\Navigation\\dist\\spa'
src_js = src_root + '\\js'
src_css = src_root + '\\css'
src_fonts = src_root + '\\fonts'
src_icons = src_root + '\\icons'
src_index = src_root + '\\index.html'

dst_root = 'C:\\Users\\16172\\Documents\\notfound945.github.io'
dst_js = dst_root + '\\js'
dst_css = dst_root + '\\css'
dst_fonts = dst_root + '\\fonts'
dst_icons = dst_root + '\\icons'
dst_index = dst_root + '\\index.html'

try:
    shutil.copytree(src_js, dst_js)
    shutil.copytree(src_css, dst_css)
    shutil.copytree(src_fonts, dst_fonts)
    shutil.copytree(src_icons, dst_icons)
    shutil.copy2(src_index, dst_index)
except shutil.Error as e:
    for src, dst, msg in e.args[0]:
         # src is source name
         # dst is destination name
         # msg is error message from exception
         print(dst, src, msg)
```
